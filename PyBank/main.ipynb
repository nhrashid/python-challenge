```python
# Import Dependencies
import pandas as pd
import os
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/budget_data.csv"
data_df = pd.read_csv(csv_path)

data_df
```

```python
# Print to terminal
str2 = "Financial Analysis"
print(str2)  
str3 = "----------------------------"
print(str3)
str4 = "Total Months:" + data_df["Date"].count().astype(str)
print(str4)
str5 = "Total: " + data_df["Profit/Losses"].astype(float).sum().astype(str)
print(str5)
max = data_df["Profit/Losses"].astype(float).max()
df = data_df["Profit/Losses"] == max

str6= "Greatest Increase in Profits:" + data_df[df]['Date'].iloc[0]+" " +data_df["Profit/Losses"].astype(float).max().astype(str)
print(str6)

min = data_df["Profit/Losses"].astype(float).min()
df = data_df["Profit/Losses"] == min

str7 = "Greatest Decrease in Profits:"   + data_df[df]['Date'].iloc[0]+" " + data_df["Profit/Losses"].astype(float).min().astype(str)
print(str7)
```

```python
# Export Result To Text File
str1 = str2+"\n"+str3+"\n"+str4+"\n"+str5+"\n"+str6+"\n" + str7+"\n"
print(str1)
#data_df.to_csv("Resources/fileOne.csv", index=False, header=True)
```

```python
# Export to Text File
base_filename = 'Values.txt'
with open(base_filename,'w') as outfile:
    outfile.write(str1)
```
